```python
cols = ['precipitation','mean_temp','min_temp','max_temp','snow_depth']

df = df[cols]
df.head()
```

```python
min_temps = df['min_temp'].idxmin()
df.loc[min_temps]
```

```python
prec, mean_temp, min_temp, max_temp, snow_dep = df.loc[min_temps]
```

```python
print(f"The lowest temperature of {min_temp} degrees occurred on {min_temps.strftime('%B %d, %Y')}, which was a {min_temps.day_name()}.")
```

```python
prec, mean_temp, min_temp, max_temp, snow_dep = df.loc[max_prec]
```
